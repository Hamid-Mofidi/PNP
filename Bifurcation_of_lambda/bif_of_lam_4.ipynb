This code is for the bifurcation analysis of flux ratio for studying the
effects of permanent charges on ionic flows

```python
from scipy.optimize import root
import numpy as np
import matplotlib.pyplot as plt
```

```python
def equations(x):
    A, l, V, I = x

    k = 1
    r = 2

    sqA    = np.sqrt(1+A**2)
    sqB    = np.sqrt(1+(l-A+r)**2) #assume that alpha=1/3, beta=2/3
    logAB  = np.log(np.nan_to_num((l-A+r)*(sqA-1), nan = +1e-15)/np.nan_to_num(A*(sqB-1), nan = +1e-15) )
    rho    = (A-l)**2 + (sqA - sqB)*(A-l)
    sigma  = l/r
    logSBA = np.log(np.nan_to_num( sigma*(l-A+r) , nan = +1e-15)/ np.nan_to_num(A, nan = +1e-15) ) # not a real number
    siglog = 3*sigma*np.log( np.nan_to_num(sigma, nan = +1e-15) )/(l*(sigma-1))

    numI1   = (( (-1)**(k+1) )*siglog) * (A-l)**2
    numI2   = (logAB - ((-1)**k) * np.log( np.nan_to_num(sigma, nan = +1e-15) ))*(A-l)
    gamma1  = 1/(I-(A-l) * sqA)
    gamma2  = 1/(I-(A-l) * sqB)
    M       = I * (gamma2 - gamma1)+ rho/I
    ABGamma = A * gamma1 + (l-A+r) * gamma2
    ABGammaR= (1/A) * gamma1 + (1/(l-A+r)) * gamma2
    G41     = ( 1- (A-l)*ABGamma )*( logSBA + (A-l) * siglog )
    G42     = ( 1- (A-l)*ABGamma )* M
    G43     = ( I+ ( ((-1)**k) * (A-l) ) ) * siglog * M/(A-l)
    GR      =( ( I**2-(A-l)**2 )/(A-l) ) * M * ABGammaR



    f1 = rho - (I* ( np.nan_to_num(np.log(np.nan_to_num( I-(A-l)*sqB, nan = +1e-10) /np.nan_to_num( I-(A-l)*sqA , nan = +1e-10) ) , nan = +1e-10) ))
    f2 = V - logAB + ((I*logSBA-rho)/(A-l))
    f3 = I - ( (numI1 + numI2+rho)/(logSBA + siglog*(A-l))  )
    f4 = G41 - G42 + G43 - GR

    return [f1, f2, f3, f4]

```

```python
# set a range of initial values
A_range = np.linspace(0, 2, 10)
l_range = np.linspace(1, 2, 10)
V_range = np.linspace(-80, 80, 10)
I_range = np.linspace(-60, 60, 10)


r = 2
# Create a boolean mask for the constraint
# The resulting arrays for A_range, I_range, V_range, and L_range
# will only contain values that satisfy the constraint.
expression1 = (A_range - l_range) * np.sqrt(1 + A_range**2)
expression2 = (A_range - l_range) * np.sqrt(1 + (l_range - A_range + r)**2)

max_expression = np.maximum(expression1, expression2)
min_expression = np.minimum(expression1, expression2)

mask = np.logical_or(I_range > max_expression, I_range < min_expression)

# Apply the mask to all arrays
A_range = A_range[mask]
l_range = l_range[mask]
V_range = V_range[mask]
I_range = I_range[mask]



# loop through all combinations of initial values
# Create an empty matrix to store solutions
sol_mtrx = np.empty((4, 0))

for a in A_range:
    for ll in l_range:
        for v in V_range:
            for i in I_range:
                x0 = [a, ll, v, i]
                sol = root(equations, x0, method='hybr')
                if sol.success:
                    # Append the solution vector to the solution matrix
                    sol_mtrx = np.append(sol_mtrx, np.array([sol.x]).T, axis=1)

# solution_matrix is initialized as an empty matrix with dimensions (4, 0). Inside the nested loops,
# for every successful solution sol, sol.x (which is a vector of length 4) is appended as a new column
# to solution_matrix using np.append along the axis 1.
# By the end of the loop, solution_matrix will have dimensions (4, n), where each row corresponds to
# the solutions of A_sol, l_sol, V_sol, and I_sol, respectively.
```

```python
A_sol = sol_mtrx[0, :]
l_sol = sol_mtrx[1, :]
V_sol = sol_mtrx[2, :]
I_sol = sol_mtrx[3, :]
print(np.shape(sol_mtrx))
```
