```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Data
L_R_pairs = [(1, 0.9), (1, 1.1), (1, 2), (1.1, 1)]
V_Q_pairs = [(-1.80, 1.50), (-1.81, 1.21), (-1.85, 0.62), (-1.82, 1.48)]

# Extracting L, R, V*, and Q* for plotting
L_values = [pair[0] for pair in L_R_pairs]
R_values = [pair[1] for pair in L_R_pairs]
V_star = [pair[0] for pair in V_Q_pairs]
Q_star = [pair[1] for pair in V_Q_pairs]

# Creating the plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plotting the data
scatter = ax.scatter(L_values, R_values, Q_star, c=V_star, cmap='viridis', s=100)

# Adding a color bar to indicate the values of V*
cbar = plt.colorbar(scatter, ax=ax, pad=0.1)
cbar.set_label('$V^*$')

# Adding titles and labels
ax.set_title('3D Scatter Plot of $(L, R, Q^*)$ with Color Indicating $V^*$')
ax.set_xlabel('$L$')
ax.set_ylabel('$R$')
ax.set_zlabel('$Q^*$')

# Show the plot
plt.show()
```
